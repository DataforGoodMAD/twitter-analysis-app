```python
import os
import spacy
import pandas as pd
import tweepy
from dotenv import load_dotenv
load_dotenv()
from collections import defaultdict
consumer_key = os.environ.get('CONSUMER_KEY')
consumer_secret = os.environ.get('CONSUMER_SECRET_KEY')
auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
df = pd.read_csv(f'./output/timeline_helsinkiespana.csv')
df.head()
```

```python
nlp = spacy.load("es_core_news_md")
```

```python
with nlp.disable_pipes(nlp.pipe_names):
    docs = [nlp(doc) for doc in df['full_text']]
```

```python
import random
docs_sample = random.sample(docs, 100)
```

```python
def requestTimeline(username, number=200, include_rts=False, exclude_replies=True, since_id=None):
    tweets = tweepy.Cursor(api.user_timeline, id=username, tweet_mode='extended', include_rts=include_rts, exclude_replies=exclude_replies, since_id=since_id).items(number)
    tweets_lst = [tweet._json for tweet in tweets]
    print(f'request for {username} suceeded')
    return tweets_lst

def requestFollowers(username, number=200):
    """
    number stands for the max number of users to retrieve. It's set at 200 for security.
    Each user is a request, and twitter allow only 3200 requests every 15 minutes. Set to None to retrieve all followers.
    """
    followers = tweepy.Cursor(api.followers, id=username, count=200).items(number)
    followers_lst = [follower._json for follower in followers]
    return followers_lst
    
    
#    df_followers.to_csv(f'./output/followers_helsinkiespana.csv', index=False)
#    df_followers.head()
```

```python
fol_10 = requestFollowers("melindasharon", 20)
```

```python
tweets2ndGrade = defaultdict(list)
for e in fol_10:
    tweets2ndGrade[e['screen_name']] = requestTimeline(username=e['screen_name'], number=20)
```

```python
users_docs = defaultdict(list)
for user, tweets in tweets2ndGrade.items():
    users_docs[user] = [nlp(tweet['full_text']) for tweet in tweets]    
        
```

```python
similarity = defaultdict(list)
for user, comps in users_docs.items():
    similarity[user] = [comp.similarity(doc) for comp in comps
                    for doc in docs_sample]
```
