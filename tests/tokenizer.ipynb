```python
punctuation = string.punctuation + "¡"
stop_words = set(stopwords.words('spanish') + stopwords.words('english'))
print(len(stop_words))
bad_words = set(['RT', 'rt'])
stop_words.update(bad_words)
print(len(stop_words))
print(type(stop_words))
'RT' in stop_words
```

```python
punctuation = '¡!"$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
```

```python


def preprocessing(tweet, stop_words):
    """
    Make lowercase, remove punctuation, stopwords & urls, lemmatize
    
    """
    lemmatizer = WordNetLemmatizer()
    tweet = re.sub(r'http\S*', '', tweet)
    tweet = unicodedata.normalize(
            'NFKD', tweet).encode('ASCII', 'ignore').decode('utf-8')
    words_list = str(tweet).translate(str.maketrans('', '', punctuation)).lower().split()
    token_list = [lemmatizer.lemmatize(word) for word in words_list if word not in stop_words and not word.isnumeric() and len(word) > 2]
    print(words_list)
    print(token_list)
    return token_list

# tokenize
# remove stopwords
# remove accent
# make lowercase
# lemmatize

#preprocessing pipeline
```

```python
counter = Counter()
def update_counter(tweet):
    counter.update(tweet)
```

```python
for tweet_text in df['full_text']:
    tweet_tokens = preprocessing(tweet_text, stop_words)
    counter.update(tweet_tokens)
```

```python
counter.most_common()
```

```python
"".join(list(map(lambda x : f'{x[0]} OR ', counter.most_common()))[0:20])[0:-4]
```

```python
hashtags = list(filter(lambda x: re.match(r"^#", x[0]), counter.most_common()))
```
